## Expresiones condicionales (if /else / ifelse) ##

R usa las palabras clave if/else. La condición a comprobar se encierra entre paréntesis y se evalua como verdadera o falsa. En función de esto se ejecuta el código correspondiente que va encerrado entre llaves.

    if(condición){
        codigo si la condición se evalúa como TRUE
        } else {
            codigo si la condición se evalúa como FALSE
        }

```r
if(a!=0){
    print(1/a)
} else {
    print("error en la operación: división por cero")
}
```

la palabra clave **ifelse** permite escribir sentencias condicionales muy compactas. Su estructura es

    ifelse(condición, codigo si verdadero, codigo si falso)

```r
a <- 0
ifelse(a>0, 1/a, NA)
```

Esta función se puede aplicar a un vector y el resultado es otro vector

```r
a <- c(0, 1, 2, -4, 5)
division  <- ifelse(a>0, 1/a, NA)
division
```

Podemos usarla para sustituir los valores NA por el valor 0 (cero)

```r
na_serie <- c(NA, 1, 5, NA, 7, 9)
sum(is.na(na_serie))
no_nas <- ifelse(is.na(na_serie), 0, na_serie)
no_nas
```

## any, all ##

**any** devuelve TRUE si al menos uno de los valores lógicos de un vector es TRUE

**all** devuelve TRUE si todos los valores lógicos de un vector son TRUE

```r
z <- c(TRUE, TRUE, FALSE)
any(z)
all(z)
any(!z)
all(!z)
```

## Funciones ##

Las funciones son objetos que se asignan mediante la palabra clave **function** y los argumentos

    una_funcion <- function(x,y,z){
        ...
        código
        ...
        valor o vector devuelto
    }

```r
# crea una funcion que devuelve la suma
# de los cuadrados de una serie
compute_s_n <- function(n){
    x <- 1:n
    sum(x^2)
}
compute_s_n(100)
```

## bucles For ##

    for (i in range of values){
        ...
        código
        ...
    }

```r
# crea un vector con la suma de los cuadrados de los primeros 25 numeros

s_n <- vector("numeric", 25) # vector vacío de longitud 25

for (i in 1:25){
    s_n[i] <- compute_s_n(i)
}
s_n
```
